### Q1 Which team has won/lost maximum number of matches while chasing? 

```python
df.query("toss_decision=='field' & toss_winner==winner | (toss_decision=='bat' & team2 == winner)")['winner'].value_counts()
```

### Q2 Which toss decision is best for each ground?


```python
team1 = df.query("team1 == winner")
team2 = df.query("team2 == winner")
grounds = team1['venue'].value_counts().index
# grounds

for ground in grounds:
    if(team1['venue'].value_counts()[ground] > team2['venue'].value_counts()[ground]):
        print(ground + '\t Bat')
    else:
        print(ground + '\t Field')
```
